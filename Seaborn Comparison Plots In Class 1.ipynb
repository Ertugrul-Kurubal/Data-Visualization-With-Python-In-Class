### *Grids*

### *1. Catplot*

```python
sns.catplot(x="sex", y="total_bill",data=tips);
```

```python
sns.catplot(x="sex", y="total_bill",data=tips, kind="box");
```

```python
sns.catplot(x="sex", y="total_bill",data=tips, kind="box", row="time");
```

```python
sns.catplot(x="sex", y="total_bill",data=tips, kind="box", col="time");
```

### *2. Pair Grid*

```python
g=sns.PairGrid(tips)
g.map(plt.scatter);
```

```python
g=sns.PairGrid(tips)
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color="r");
```

```python
g=sns.PairGrid(mpg)
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color="r");
```

### *3. FacetGrid*

```python
g=sns.FacetGrid(tips, col="time")
g.map(sns.kdeplot,"total_bill");
```

```python
g=sns.FacetGrid(tips, row="smoker")
g.map(sns.kdeplot,"total_bill");
```

```python
g=sns.FacetGrid(tips, col="time",row="smoker")
g.map(sns.kdeplot,"total_bill");
```

```python
plt.figure(figsize=(12,7))
sns.pointplot(x="vehicleType", y="price", data=df, order=df.groupby("vehicleType")["price"].mean().sort_values(ascending=False).index);
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x="vehicleType", data=df);
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
g=sns.countplot(x="vehicleType", order=df.vehicleType.value_counts().index, data=df)
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
g=sns.countplot(x="vehicleType", order=df.vehicleType.value_counts().index, data=df)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.2, p.get_height()+1000))
```
