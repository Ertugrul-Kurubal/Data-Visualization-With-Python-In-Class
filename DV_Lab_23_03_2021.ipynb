```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,30)
y = x*2
z = x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('OO other method')
ax.plot(x,y,'green',ls='--',lw=5);
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2, figsize=(14,6))

ax[0].plot(x,y,'blue',ls='--',label='x vs y')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].spines['left'].set_color('blue')
ax[0].set_xticks(x)
ax[0].set_xticklabels(x,rotation=90)
ax[0].legend()

ax[1].plot(x,z,'r',ls='-.',label='x vs z',marker='+',markersize=10,markevery=5)
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
ax[1].legend(loc=(1.1,0.5))

plt.show()
```

**Pie Chart**

```python
fig,ax=plt.subplots(figsize=(14,6))

data = [23, 17, 35, 29, 12, 41]

cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES']

wp={'linewidth':1,'edgecolor':'green'}
my_labels=cars
my_colors=['orange','red','cyan','grey','green','indigo']
my_explode=(0,0,0.2,0,0.4,0)

ax.pie(data,labels=my_labels,autopct='%1.1f%%',explode=my_explode,startangle=90,wedgeprops=wp,shadow=True)
ax.set_title('Cars')
ax.axis('equal')
ax.legend(loc=('center right'))
plt.show()
```

**Box Plot**

```python
data = pd.read_csv('hist_data.csv')
```

```python
data.describe().T
```

```python
data.head()
```

```python
ages=data['Age']
```

```python
len(data)
```

```python
from scipy.stats import iqr
```

```python
q75,q25=np.percentile(ages ,[75,25])
```

```python
lower_bound=q25-(1.5*iqr(ages))
```

```python
upper_bound=q75+(1.5*iqr(ages))
```

```python
print("lower_bound:", lower_bound)
print("upper_bound:", upper_bound)
```

```python
print("min_value:", min(ages))
print("max_vale:", max(ages))
```

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.boxplot(ages)
ax.axhline(24)
ax.axvline(1.1)
ax.axvline(0.9)
ax.axhline(35);
```

**Bar Plot**

```python
teams = ['GS', 'FB', 'BJK', 'TS', 'ALN']
points = [61, 59, 64, 55, 49]
```

```python
fig,ax=plt.subplots(figsize=(14,5))
ax.bar(teams,points, width=0.8,color=['red','blue','black','purple','orange'])
plt.suptitle('LeaderShip')
ax.set_title('Eagle')
ax.set_xlabel('Teams')
ax.set_ylabel('Points')
ax.set_xticks(teams)
ax.set_xticklabels(teams,rotation=45)
plt.tight_layout()
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.35, p.get_height()+0.5));
plt.show()
```
