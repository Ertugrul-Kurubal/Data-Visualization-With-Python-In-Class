```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
```

```python
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

## *Matplotlib Functional*

```python
plt.plot(age,salary,"g");
```

```python
plt.plot(age,salary,"g")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
;
```

```python
plt.plot(age,salary,"g",label="salary")
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend();
```

```python
plt.plot(age,salary,"g",label="salary")
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend(loc=4);
```

```python
plt.plot(age,salary,"g",label="salary")
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend(loc=1);
```

```python
plt.subplot(1,2,1)
plt.plot(age,salary,"g",label="salary")
plt.subplot(1,2,2)
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend()
plt.tight_layout();
```

```python
plt.subplot(1,2,1)
plt.plot(age,salary,"g",label="salary")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend()
plt.subplot(1,2,2)
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend()
plt.tight_layout();
```

## *Matplotlib Object-Oriented*

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"b")
ax.set_ylabel("salary")
ax.set_xlabel("age")
ax.set_title("Age vs Salary");
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age,salary,"r")
ax[0,0].set_title("Age vs Salary")
ax[0,1].plot(age,salary_2,"b")
ax[0,1].set_title("Age vs Salary2")
ax[1,0].plot(age,salary_2,"g")
ax[1,0].set_title("Age vs Salary2")
ax[1,1].plot(age,salary,"r")
ax[1,1].set_title("Age vs Salary")
plt.tight_layout();
```

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"r")
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(x=36,color="green")
ax.axhline(y=60000, ls='--')
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Age vs Salary")
```

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"r",label="salary")
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(x=36,color="green")
ax.axhline(y=60000, ls='--')
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Age vs Salary")
plt.legend(loc=4) # ax.legend
fig.savefig("hatıra.png",dpi=150);
```

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"r",label="salary")
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(x=36,color="green")
ax.axhline(y=60000, ls='--')
ax.set_xticks(age)
ax.set_xticklabels(age)
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Age vs Salary")
plt.legend(loc=(1.1,0.5));
```
