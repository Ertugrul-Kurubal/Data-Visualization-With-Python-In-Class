## *Matplotlib Object-Oriented*

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"b")
ax.set_ylabel("salary")
ax.set_xlabel("age")
ax.set_title("Age vs Salary");
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age,salary,"r")
ax[0,0].set_title("Age vs Salary")
ax[0,1].plot(age,salary_2,"b")
ax[0,1].set_title("Age vs Salary2")
ax[1,0].plot(age,salary_2,"g")
ax[1,0].set_title("Age vs Salary2")
ax[1,1].plot(age,salary,"r")
ax[1,1].set_title("Age vs Salary")
plt.tight_layout();
```

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"r")
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(x=36,color="green")
ax.axhline(y=60000, ls='--')
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Age vs Salary")
```

```python
fig,ax = plt.subplots()
ax.plot(age,salary,"r",label="salary")
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(x=36,color="green")
ax.axhline(y=60000, ls='--')
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Age vs Salary")
plt.legend(loc=4) # ax.legend
fig.savefig("hatıra.png",dpi=150);
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
langs = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs);
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs, autopct="%1.2f%%");
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs, autopct="%1.2f%%", explode=(0,0,0.1,0,0));
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs, autopct="%1.2f%%",startangle=90, explode = explode )
explode=(0,0,0.3,0,0);
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs, autopct="%1.2f%%",startangle=45, explode = explode )
explode=(0,0,0.3,0,0);
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs, autopct="%1.2f%%",startangle=45, explode = explode,shadow=True)
explode=(0,0,0.3,0,0);
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
ax.pie(slices, labels=langs, autopct="%1.2f%%",startangle=45, explode = explode,shadow=True)
explode=(0,0,0.3,0,0)
ax.axis("equal");
```

```python
fig ,ax = plt.subplots(figsize=(10,6))
explode=(0.1,0,0.3,0.5,0)
ax.pie(slices, labels=langs, autopct="%1.1f%%", explode = explode,shadow=True)
ax.axis("equal");
```
