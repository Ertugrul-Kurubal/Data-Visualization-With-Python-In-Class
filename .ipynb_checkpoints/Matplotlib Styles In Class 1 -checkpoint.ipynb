```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
```

```python
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age,salary,"g");
```

```python
plt.plot(age,salary,"g")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
;
```

```python
plt.plot(age,salary,"g",label="salary")
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend();
```

```python
plt.plot(age,salary,"g",label="salary")
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend(loc=4);
```

```python
plt.plot(age,salary,"g",label="salary")
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend(loc=1);
```

```python
plt.subplot(1,2,1)
plt.plot(age,salary,"g",label="salary")
plt.subplot(1,2,2)
plt.plot(age,salary_2,"r",label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age vs Salary")
plt.legend()
plt.tight_layout();
```
